```python
import pandas as pd
```

```python
ds = pd.read_csv('kp250.csv')
dost = pd.read_csv('dost.csv')

df = pd.concat([ds,dost],axis=1)
df
```

```python
df.to_csv('ds_with_dost.csv')
```

```python
ussr = df.loc[df['country']=='СССР']
rus = df.loc[df['country']=='Россия']
rus = pd.concat([ussr,rus])
rus
```

```python
import matplotlib.pyplot as plt

ax = rus.plot(kind='scatter', x = 'year', y='positive', figsize=(10, 6), color='k')
ax.set_facecolor("oldlace")
plt.title('Позитивное настроение в фильмах СССР И России из топ 250 Кинопоиска')
```

```python
rusp= rus.loc[rus['positive']>0.1]

ax = rusp.plot(kind='scatter', x = 'year', y='positive', figsize=(10, 6), color='k')
ax.set_facecolor("oldlace")
plt.title('Позитивноее в фильмах СССР и России из топ 250 Кинопоиска')
```

```python
import pandas as pd

df = pd.read_csv('kp250.csv')
```

```python
df = df[['movie','year','country', 'rating_ball']]
df['country'].astype(str)

# Make the plot with pandas
df['country'].value_counts()
```

```python
 import matplotlib.pyplot as plt 
    
df = df['country'].value_counts()

# Make the plot with pandas
df.plot(kind='pie', subplots=True, figsize=(15, 15))
plt.text(
    0.1, 1.2,
    "Топ 250 Кинопоиска по странам",
    size=25,
    ha="center",color="#000000"
)

plt.ylabel("")
plt.show()
```
